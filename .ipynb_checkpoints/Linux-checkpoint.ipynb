# 一.Docker 安裝

## Step1：更新系统软件包
sudo apt update

## Step2：安装依赖包【用于通过HTTPS来获取仓库】
sudo apt install apt-transport-https ca-certificates curl software-properties-common

## Step3：添加Docker官方GPG密钥
sudo -i
curl -fsSL https://mirrors.aliyun.com/docker-ce/linux/ubuntu/gpg | gpg --dearmor -o /etc/apt/trusted.gpg.d/docker-ce.gpg

## Step4：验证
sudo apt-key fingerprint 0EBFCD88
0EBFCD88 是公钥的指纹。执行这个命令后，系统会显示与该指纹相关的公钥信息。

## Step4：添加Docker阿里稳定版软件源
sudo add-apt-repository "deb [arch=amd64] https://mirrors.aliyun.com/docker-ce/linux/ubuntu $(lsb_release -cs) stable"

## Step5：再次更新软件包
sudo apt update

## Step6：安装默认最新版
sudo apt install docker-ce docker-ce-cli containerd.io
查看状态：sudo systemctl status docker
启动：sudo systemctl start docker
开机自启：sudo systemctl enable docker
停止：sudo systemctl stop docker

## （八）其它常用基本命令
### 1. rm - 删除文件或目录
#### 功能：删除文件或目录（慎用！不可逆操作）
#### 常用选项：

####  -r 或 -R：递归删除目录及其内容

####  -f：强制删除，忽略不存在的文件或警告

####  -i：交互式删除，删除前确认

#### 示例：


#### rm file.txt            # 删除文件
#### rm -r my_folder        # 递归删除目录（包含子目录和文件）
#### rm -rf /path/to/dir    # 强制删除目录（危险！可能误删系统文件）
#### rm -i *.log            # 交互式删除所有.log文件（逐个确认）

### 2. which - 查找命令的绝对路径
#### 功能：在 PATH 环境变量中查找可执行文件的路径
#### 示例：

#### which python          # 输出: /usr/bin/python（显示第一个匹配的路径）
#### which -a python       # 显示所有匹配的路径（如不同版本的Python）

### 3. find - 搜索文件或目录
#### 功能：根据名称、类型、时间等条件递归搜索文件
#### 常用选项：

#### -name：按文件名匹配（支持通配符 *）

#### -type：按类型筛选（f 文件，d 目录）

#### -mtime：按修改时间筛选（如 -mtime +7 表示7天前）

#### -exec：对搜索结果执行命令

#### 示例：


#### find /home -name "*.txt"        # 在/home下搜索所有.txt文件
#### find . -type d                  # 查找当前目录下所有子目录
#### find /var/log -mtime -1        # 查找过去24小时内修改过的文件
#### find . -size +10M              # 查找大于10MB的文件
#### find /tmp -name "*.tmp" -exec rm {} \;  # 删除/tmp下所有.tmp文件

### 4. grep - 文本搜索工具
#### 功能：在文件或输入流中匹配正则表达式
#### 常用选项：

#### -i：忽略大小写

#### -r 或 -R：递归搜索目录中的文件

#### -n：显示匹配行的行号

#### -v：反向匹配（显示不包含模式的行）

#### -E：支持扩展正则表达式（等同于 egrep）

#### 示例：


#### grep "error" log.txt           # 在log.txt中查找包含"error"的行
#### grep -r "404" /var/log/nginx   # 递归搜索nginx目录中所有包含"404"的文件
#### grep -i "warning" system.log   # 忽略大小写搜索"warning"
#### grep -n "func(" app.py         # 显示匹配行的行号
#### echo "hello" | grep "he"       # 从管道输入中匹配内容
### 5. wc - 统计行数、单词数、字符数
#### 功能：统计文件或输入的行数（-l）、单词数（-w）、字符数（-c）
#### 示例：


#### wc file.txt            # 输出：行数 单词数 字符数 文件名
#### wc -l access.log      # 统计文件的行数（常用！）
#### ls | wc -w            # 统计当前目录下文件/目录的数量（单词数）
#### echo "hello world" | wc -c  # 统计字符数（包括换行符，输出: 12）
### 6. echo - 输出文本
#### 功能：输出文本到终端或文件（结合重定向）
#### 常用选项：

#### -e：启用转义字符（如 \n 换行）

#### -n：不自动添加换行符

#### 示例：


#### echo "Hello World"             # 输出文本
#### echo -e "Line1\nLine2"         # 输出两行（启用转义）
#### echo -n "No newline" > file.txt # 写入文件时不换行
#### echo $PATH                     # 输出环境变量内容


### 7. tail - 查看文件尾部内容
#### 功能：查看文件的末尾部分（默认10行），常用于实时监控日志
#### 常用选项：

#### -n：指定显示的行数（如 -n 20）

#### -f：实时追踪文件更新（常用于监控日志）

#### -F：类似 -f，但会跟踪文件重命名或重建

#### 示例：

#### tail log.txt           # 显示文件最后10行
#### tail -n 20 log.txt     # 显示最后20行
#### tail -f /var/log/syslog # 实时监控日志（按 Ctrl+C 退出）
#### tail -F app.log        # 持续跟踪，即使日志文件被轮转

#### 7.使用 ls -l 查看权限
每行包含 7 个字段，从左到右依次为：
权限 链接数 所有者 所属组 大小 修改时间 名称
权限字段格式：-|d|l rwx rwx rwx
共 10 个字符，分为 4 部分：

第 1 个字符：文件类型

d：目录（如 a, Desktop）

-：普通文件（如 ada, ada2）

l：符号链接（未在示例中出现）

后续 9 个字符：权限分三组（所有者、所属组、其他用户）

r = 读权限（值 4）

w = 写权限（值 2）

x = 执行权限（值 1）

#### - = 无对应权限

示例解析
drwxr-xr-x（目录 a 的权限）

所有者权限：rwx（可读、写、执行）

组权限：r-x（可读、执行，不可写）

其他用户权限：r-x（同上）

-rw-r--r--（文件 ada 的权限）

所有者权限：rw-（可读、写，不可执行）

组和其他用户权限：r--（仅可读）
#### 8.chmod 修改文件权限
#### 格式：chmod [用户类型][操作符][权限] 文件

用户类型：u（所有者）、g（组）、o（其他）、a（全部）

操作符：+（添加）、-（移除）、=（直接设置）

权限：r、w、x
chmod u=rwx,g=rx,o-x test3.txt,将文件权限修改为rwxr-x--x

u代表用户权限，g为用户组权限，o为other其它用户权限
